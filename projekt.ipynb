Średnia cena gier wykazuje pewne fluktuacje, ale ogólnie można zauważyć pewne spadki cen w ostatnich latach. Może to być wynikiem zwiększonej konkurencji na rynku lub wzrostu popularności modeli free-to-play, co wpływa na średnią cenę.

Ogólny trend w popularności tagów — dla większości wybranych tagów (takich jak RPG, Strategiczne, Symulatory, Akcja, Wieloosobowe) można zauważyć rosnące zainteresowanie, co odzwierciedla rozwój i zmiany preferencji graczy na przestrzeni lat. 
Dekompozycja dla tagu "Strategiczne" — pokazuje wyraźny wzrost popularności tego gatunku, zwłaszcza w ostatnich latach.

Heatmapa współwystępowania tagów prezentuje, które tagi najczęściej występują razem. Silne korelacje wskazują na tagi, które często są używane razem w grach. Na przykład, jeśli tagi "Otwarty świat" i "RPG" często współwystępują, może to sugerować, że gry z otwartym światem często posiadają elementy RPG, co jest atrakcyjne dla graczy. 
Dalsze analizy mogłyby obejmować dokładniejsze badanie, jak te kombinacje wpływają na sukces gry, oceniany na podstawie liczby recenzji lub średniej oceny.

Analizując zależność między ceną a liczbą recenzji, możemy zrozumieć, jak cena wpływa na decyzje zakupowe graczy i jak postrzegana wartość (nacechowanie recenzji) jest związane z popularnością gry. Gry z umiarkowanymi cenami i pozytywnym nacechowaniem recenzji zdobywają więcej uwagi i są częściej kupowane, co sugeruje, że stosunek jakości do ceny jest ważny dla sukcesu.
